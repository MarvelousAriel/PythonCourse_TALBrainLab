```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import rc
rc('mathtext', default='regular')
```

```python
# 小数不用科学计数法显示
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
ic_data = pd.read_csv('解析后日志数据\解析后日志数据\df_arranged_注意抑制训练.csv')
```

```python
ic_data
```

# 选一位幸运用户

```python
luckyguy = ic_data[(ic_data.userid == 56144233) & (ic_data.interactionid == 3215261131510255666 )]
luckyguy
```

# 某一讲的数据

```python
plt.plot(luckyguy['num'], luckyguy['accuracy'])
new_ticks = np.linspace(1, 18, 18)
plt.xticks(new_ticks)
plt.xlabel("trial")
plt.ylabel("ACC")
plt.show()
```

# 正确率和时间数据放在一起

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(luckyguy['num'][luckyguy.isempty == 0], luckyguy['accuracy'][luckyguy.isempty == 0], c = 'red', label = 'ACC')
plt.xticks(np.linspace(1, 18, 18))
plt.yticks(np.linspace(0.5, 1, 11))
plt.xlabel("trial")
ax.set_ylabel('ACC',color='red')
ax2 = ax.twinx()
ax2.plot(luckyguy['num'][luckyguy.isempty == 0], luckyguy['time'][luckyguy.isempty == 0], c = 'orange', label = 'RT')
plt.yticks(np.linspace(400, 2500, 5))
plt.xticks(np.linspace(1, 18, 18))
plt.xlabel("trial")
ax2.set_ylabel('RT',color='orange')
plt.axvline(x = 9.5, ls = "-", c = "green")
```

# 各讲次的数据

```python
# 筛选我们想看的trial
luckyguy_return = ic_data[(ic_data.userid == 56144233) & (ic_data.isempty == 0)]
luckyguy_return
```

# 各讲次平均

```python
multi_class = luckyguy_return.groupby(['num']).mean()
multi_class
# multi_class.index
# multi_class['accuracy']['mean']
```

```python
fig_accum = plt.figure()
ax = fig_accum.add_subplot()
ax.plot(multi_class.index, multi_class['accuracy'], c = 'red')
plt.xticks(np.linspace(1, 22, 22))
plt.yticks(np.linspace(0.5, 1, 11))
plt.xlabel("trial")
ax.set_ylabel('avgACC',color='red')
ax2 = ax.twinx()
ax2.plot(multi_class.index, multi_class['time'], c = 'orange')
plt.xticks(np.linspace(1, 22, 22))
plt.yticks(np.linspace(400, 2500, 5))
plt.xlabel("trial")
ax2.set_ylabel('avgRT',color='orange')
plt.axvline(x = 9.5, ls = "-", c = "green")
```

# 群体在一个讲次的表现

```python
# 因为有强制收题，所以只保留到17题
whole = ic_data[(ic_data.interactionid == 3215261131510255666 ) & (ic_data.num <= 17)]
whole
```

```python
whole_one = whole.groupby(['num']).mean()
whole_one
```

```python
fig_whole = plt.figure()
ax = fig_whole.add_subplot()
ax.plot(whole_one.index, whole_one['accuracy'], c = 'red')
plt.xticks(np.linspace(1, 17, 17))
plt.yticks(np.linspace(0.5, 1, 11))
plt.xlabel("trial")
ax.set_ylabel('avgACC',color='red')
ax2 = ax.twinx()
ax2.plot(whole_one.index, whole_one['time'], c = 'orange')
plt.xticks(np.linspace(1, 17, 17))
plt.yticks(np.linspace(400, 2500, 5))
plt.xlabel("trial")
ax2.set_ylabel('avgRT',color='orange')
plt.axvline(x = 9.5, ls = "-", c = "green")
```
